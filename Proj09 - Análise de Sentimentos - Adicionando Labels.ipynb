# **Adicionando Labels as Frases**

**Objetivo**

* Classificar as frases em positivas, negativas e neutras

```python
# Carregando bibliotecas

import nltk
import pandas as pd

from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
# Realizando download do Vader

nltk.download("vader_lexicon")
```

```python
# Carregando base de dados

dados = pd.read_csv('/content/drive/MyDrive/Ciência de Dados com Python/Projetos/Dados/Review_Dataset.csv')
```

```python
# Visualizando dados

dados.head()
```

```python
# Instânciando objeto

sentimentos = SentimentIntensityAnalyzer()
```

```python
# Coletando informações sentimentais sobre as frases

dados["Positivo"] = [sentimentos.polarity_scores(i)["pos"] for i in dados["Review"]]
dados["Negativo"] = [sentimentos.polarity_scores(i)["neg"] for i in dados["Review"]]
dados["Neutro"] = [sentimentos.polarity_scores(i)["neu"] for i in dados["Review"]]
dados['Compound'] = [sentimentos.polarity_scores(i)["compound"] for i in dados["Review"]]
```

```python
# Realizando Classificação

## Coletando pontuação composta das frases
score = dados["Compound"].values

## Criando uma lista vazia para salvar classificações
sentimentos = []

## Loop para percorre as pontuações compostas e realizar classificação
for i in score:
    if i >= 0.5 :
        sentimentos.append('Positivo')
    elif i < 0 :
        sentimentos.append('Negativo')
    else:
        sentimentos.append('Neutral')

## Adicionando resultados ao conjunto de dados
dados["Sentimentos"] = sentimentos
```
